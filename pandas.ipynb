```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
df["Average"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
```

```python
df.head()
```

```python
df["Average"].max()
```

```python
df["Average"].min()
```

```python
df.loc[df["Average"] >= 90, "Grade"] = "A"
df.loc[(df["Average"] >= 75) & (df["Average"] < 90), "Grade"] = "B"
df.loc[(df["Average"] >= 65) & (df["Average"] < 75), "Grade"] = "C"
df.loc[(df["Average"] >= 55) & (df["Average"] < 65), "Grade"] = "D"
df.loc[df["Average"] < 55, "Grade"] = "F"
```
